```python
import pandas as pd

# -------------------------------------
# 📌 1️⃣ Define NHANES years & cycles
# -------------------------------------
years = ["2007", "2009", "2011", "2013", "2015", "2017", "2020"]
cycles = ["E", "F", "G", "H", "I", "J", ""]

df_list = []

# -------------------------------------
# 📌 2️⃣ Loop through each cycle
# -------------------------------------
for year, cycle in zip(years, cycles):
    if year == "2020":
        prefix = "J"
        nhanes_year = "2017"
    else:
        prefix = cycle
        nhanes_year = year

    try:
        # -------------------------------------
        # Liver enzyme file (BIOPRO)
        # -------------------------------------
        liver_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/BIOPRO_{prefix}.XPT"
        df_liver = pd.read_sas(liver_url)[['SEQN', 'LBXSATSI', 'LBXSASSI', 'LBXSGTSI']]

        # -------------------------------------
        # Demographics file (DEMO) — NOW includes Race/Ethnicity!
        # ✅ FIXED: Add 'RIDRETH1'
        # -------------------------------------
        demo_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/DEMO_{prefix}.XPT"
        df_demo = pd.read_sas(demo_url)[['SEQN', 'RIDAGEYR', 'RIAGENDR', 'RIDRETH1']]

        # -------------------------------------
        # BMI file (BMX)
        # -------------------------------------
        bmi_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/BMX_{prefix}.XPT"
        df_bmi = pd.read_sas(bmi_url)[['SEQN', 'BMXBMI']]

        # -------------------------------------
        # Sleep file (SLQ)
        # -------------------------------------
        sleep_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/SLQ_{prefix}.XPT"
        df_sleep = pd.read_sas(sleep_url)
        sleep_col = 'SLD010H' if 'SLD010H' in df_sleep.columns else 'SLD012'
        df_sleep = df_sleep[['SEQN', sleep_col]].rename(columns={sleep_col: 'SleepHours'})

        # -------------------------------------
        # Alcohol file (ALQ)
        # -------------------------------------
        alcohol_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/ALQ_{prefix}.XPT"
        df_alcohol = pd.read_sas(alcohol_url)
        alcohol_col = 'ALQ101' if 'ALQ101' in df_alcohol.columns else 'ALQ111'
        df_alcohol = df_alcohol[['SEQN', alcohol_col]].rename(columns={alcohol_col: 'Alcohol'})

        # -------------------------------------
        # Physical Activity file (PAQ)
        # -------------------------------------
        pa_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/PAQ_{prefix}.XPT"
        df_pa = pd.read_sas(pa_url)[['SEQN', 'PAQ605']]

        # -------------------------------------
        # 📌 Merge all by SEQN
        # -------------------------------------
        df = (
    df_liver
    .merge(df_demo, on="SEQN", how="left")
    .merge(df_bmi, on="SEQN", how="left")
    .merge(df_sleep, on="SEQN", how="left")
    .merge(df_alcohol, on="SEQN", how="left")
    .merge(df_pa, on="SEQN", how="left")
)


        df["Cycle"] = year
        df_list.append(df)

        print(f"✅ Loaded & merged: {year}")

    except Exception as e:
        print(f"❌ Failed for {year} - {e}")

# -------------------------------------
# 📌 3️⃣ Combine all cycles
# -------------------------------------
if df_list:
    df_all = pd.concat(df_list, ignore_index=True)
    print("\n✅ Combined dataset shape:", df_all.shape)
    print(df_all["Cycle"].value_counts())
else:
    print("❌ No data merged.")


# 📊 Sample size (non-missing count) per variable
print("\n📊 Sample size (non-missing) per variable:\n")
sample_sizes = df_all.notna().sum()
print(sample_sizes)


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

years = ["2007", "2009", "2011", "2013", "2015", "2017", "2020"]
cycles = ["E", "F", "G", "H", "I", "J", ""]

df_list = []

for year, cycle in zip(years, cycles):
    if year == "2020":
        prefix = "J"
        nhanes_year = "2017"
    else:
        prefix = cycle
        nhanes_year = year

    try:
        # File URLs
        liver_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/BIOPRO_{prefix}.XPT"
        demo_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/DEMO_{prefix}.XPT"
        bmi_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/BMX_{prefix}.XPT"
        sleep_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/SLQ_{prefix}.XPT"
        alcohol_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/ALQ_{prefix}.XPT"
        pa_url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{nhanes_year}/DataFiles/PAQ_{prefix}.XPT"

        # Load datasets
        df_liver = pd.read_sas(liver_url)
        liver_cols = [c for c in ['SEQN', 'LBXSATSI', 'LBXSASSI', 'LBXSGTSI'] if c in df_liver.columns]
        df_liver = df_liver[liver_cols]

        df_demo = pd.read_sas(demo_url)[['SEQN', 'RIDAGEYR', 'RIAGENDR', 'RIDRETH1']]
        df_bmi = pd.read_sas(bmi_url)[['SEQN', 'BMXBMI']]

        df_sleep = pd.read_sas(sleep_url)
        sleep_col = 'SLD010H' if 'SLD010H' in df_sleep.columns else 'SLD012'
        df_sleep = df_sleep[['SEQN', sleep_col]].rename(columns={sleep_col: 'SleepHours'})

        df_alcohol = pd.read_sas(alcohol_url)
        alcohol_col = 'ALQ101' if 'ALQ101' in df_alcohol.columns else 'ALQ111'
        df_alcohol = df_alcohol[['SEQN', alcohol_col]].rename(columns={alcohol_col: 'Alcohol'})

        df_pa = pd.read_sas(pa_url)[['SEQN', 'PAQ605']]

        # Merge
        df = df_liver.merge(df_demo, on="SEQN", how="left") \
                     .merge(df_bmi, on="SEQN", how="left") \
                     .merge(df_sleep, on="SEQN", how="left") \
                     .merge(df_alcohol, on="SEQN", how="left") \
                     .merge(df_pa, on="SEQN", how="left")

        df["Cycle"] = year
        df_list.append(df)
        print(f"✅ Loaded: {year}")

    except Exception as e:
        print(f"❌ Failed: {year} - {e}")

# -------------------------------------
# 2) Combine all
# -------------------------------------
df_all = pd.concat(df_list, ignore_index=True)

# -------------------------------------
# 3) Rename to friendly labels
# -------------------------------------
df_all = df_all.rename(columns={
    'LBXSATSI': 'ALT (U/L)',
    'LBXSASSI': 'AST (U/L)',
    'LBXSGTSI': 'GGT (U/L)',
    'BMXBMI': 'BMI (kg/m2)',
    'SleepHours': 'Sleep (hrs)',
    'Alcohol': 'Alcohol Use',
    'PAQ605': 'Physical Activity',
    'RIDAGEYR': 'Age'
})

# -------------------------------------
# 4) Pick only numeric + clean rows
# -------------------------------------
cols = [
    'ALT (U/L)',
    'AST (U/L)',
    'GGT (U/L)',
    'BMI (kg/m2)',
    'Sleep (hrs)',
    'Alcohol Use',
    'Physical Activity',
    'Age'
]

df_heatmap = df_all[cols].dropna()

# -------------------------------------
# 5) Plot clearer heatmap with clustering
# -------------------------------------
plt.figure(figsize=(12, 10))
sns.clustermap(
    df_heatmap.corr(),
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    linewidths=0.5,
    figsize=(12, 10)
)
plt.suptitle("Correlation Heatmap of Liver Enzymes & Lifestyle Factors", fontsize=16)
plt.show()
```

```python
# -------------------------------------------------
# EDA Block: Data Exploration & Descriptives
# -------------------------------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import shapiro
from statsmodels.stats.outliers_influence import variance_inflation_factor
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from itertools import combinations

# -------------------------------------------------
# Descriptive statistics
# -------------------------------------------------
num_vars = [
    'ALT (U/L)',
    'AST (U/L)',
    'GGT (U/L)',
    'BMI (kg/m2)',
    'Sleep (hrs)',
    'Alcohol Use',
    'Physical Activity',
    'Age'
]

# ✅ Drop all rows with any NA in these columns
df_clean = df_all[num_vars].dropna()

print("Shape after dropping NA:", df_clean.shape)

# ✅ Run Shapiro-Wilk on each (if n > 5000, sample)
shapiro_results = []

for col in num_vars:
    data = df_clean[col]
    n = len(data)

    if n > 5000:
        data = data.sample(5000, random_state=42)

    w_stat, p_value = shapiro(data)
    shapiro_results.append({
        "Variable": col,
        "N": n,
        "W-Statistic": w_stat,
        "p-Value": p_value
    })

# ✅ Make a nice table
shapiro_df = pd.DataFrame(shapiro_results)
print(shapiro_df)


# ✅ Calculate fresh descriptive stats for the NA-dropped version
desc_stats = df_clean.describe().T
desc_stats['Skewness'] = df_clean.skew()
display(desc_stats)

# -------------------------------------------------
# Log transform highly skewed enzymes
# -------------------------------------------------
df_all['log_ALT'] = np.log1p(df_all['ALT (U/L)'])
df_all['log_AST'] = np.log1p(df_all['AST (U/L)'])
df_all['log_GGT'] = np.log1p(df_all['GGT (U/L)'])

# --------------------------------------------------------
# 1️⃣ T-test + Mann–Whitney + Cohen’s d
# --------------------------------------------------------

male = df_all[df_all['RIAGENDR'] == 1]['ALT (U/L)'].dropna()
female = df_all[df_all['RIAGENDR'] == 2]['ALT (U/L)'].dropna()

t_stat, p_val = stats.ttest_ind(male, female, equal_var=False)
print(f"T-test ALT Male vs Female: t = {t_stat:.2f}, p = {p_val:.4f}")

u_stat, p_val_u = stats.mannwhitneyu(male, female, alternative='two-sided')
print(f"Mann–Whitney U ALT Male vs Female: U = {u_stat:.2f}, p = {p_val_u:.4f}")

cohens_d = (male.mean() - female.mean()) / np.sqrt(
    ((male.std() ** 2) + (female.std() ** 2)) / 2
)
print(f"Cohen's d: {cohens_d:.2f} (effect size)")

# --------------------------------------------------------
# 2️⃣ ANOVA + Kruskal + Tukey post-hoc + Pairwise MWU
# --------------------------------------------------------

anova_data = df_all[['ALT (U/L)', 'RIDRETH1']].dropna()
groups = [
    anova_data['ALT (U/L)'][anova_data['RIDRETH1'] == g]
    for g in anova_data['RIDRETH1'].unique()
]

f_stat, p_val = stats.f_oneway(*groups)
print(f"ANOVA ALT by Race: F = {f_stat:.2f}, p = {p_val:.4f}")

h_stat, p_val_h = stats.kruskal(*groups)
print(f"Kruskal–Wallis ALT by Race: H = {h_stat:.2f}, p = {p_val_h:.4f}")

anova_data['Race'] = anova_data['RIDRETH1'].astype(str)
tukey = pairwise_tukeyhsd(
    endog=anova_data['ALT (U/L)'],
    groups=anova_data['Race'],
    alpha=0.05
)
print("\nTukey HSD Post-hoc:")
print(tukey)

print("\nPairwise Mann–Whitney U tests:")
for g1, g2 in combinations(anova_data['RIDRETH1'].unique(), 2):
    grp1 = anova_data['ALT (U/L)'][anova_data['RIDRETH1'] == g1]
    grp2 = anova_data['ALT (U/L)'][anova_data['RIDRETH1'] == g2]
    stat, p = stats.mannwhitneyu(grp1, grp2, alternative='two-sided')
    print(f"Group {g1} vs {g2}: U = {stat:.2f}, p = {p:.4f}")

# --------------------------------------------------------
# 3️⃣ Multiple Linear Regression + Residual Plot + VIF
# --------------------------------------------------------

model_data = df_all[[
    'ALT (U/L)', 'BMI (kg/m2)', 'Sleep (hrs)',
    'Alcohol Use', 'Physical Activity', 'Age', 'RIAGENDR'
]].dropna()

X = model_data[[
    'BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use',
    'Physical Activity', 'Age', 'RIAGENDR'
]]
X = sm.add_constant(X)
y = model_data['ALT (U/L)']

model = sm.OLS(y, X).fit()
print(model.summary())

residuals = model.resid
fitted = model.fittedvalues

sns.scatterplot(x=fitted, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted ALT')
plt.ylabel('Residuals')
plt.show()

vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["Variable"] = X.columns
print(vif)

# --------------------------------------------------------
# 4️⃣ Stratified: Males only & Females only
# --------------------------------------------------------

male_data = model_data[model_data['RIAGENDR'] == 1]
X_m = male_data[[
    'BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use',
    'Physical Activity', 'Age'
]]
X_m = sm.add_constant(X_m)
y_m = male_data['ALT (U/L)']

model_male = sm.OLS(y_m, X_m).fit()
print("\n--- Regression: Males only ---")
print(model_male.summary())

female_data = model_data[model_data['RIAGENDR'] == 2]
X_f = female_data[[
    'BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use',
    'Physical Activity', 'Age'
]]
X_f = sm.add_constant(X_f)
y_f = female_data['ALT (U/L)']

model_female = sm.OLS(y_f, X_f).fit()
print("\n--- Regression: Females only ---")
print(model_female.summary())
```

```python
# -------------------------------------------------
# 📌 Group Comparisons & Hypothesis Testing
# Purpose:
# - Compare mean liver enzyme levels between groups (e.g. Gender, Race)
# - Test for significant differences
# - Run Multiple Linear Regression to see predictors' effects.
# -------------------------------------------------

import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

# -------------------------------------------------
# 📌 1️⃣ Example t-test: ALT in Males vs Females
# Purpose: Test if mean ALT differs by gender.
# RIAGENDR: 1 = Male, 2 = Female
# -------------------------------------------------

male_ALT = df_all[df_all['RIAGENDR'] == 1]['ALT (U/L)'].dropna()
female_ALT = df_all[df_all['RIAGENDR'] == 2]['ALT (U/L)'].dropna()

t_stat, p_val = stats.ttest_ind(male_ALT, female_ALT, equal_var=False)
print(f"T-test ALT Male vs Female: t-stat = {t_stat:.2f}, p = {p_val:.4f}")

# -------------------------------------------------
# 📌 2️⃣ Example ANOVA: ALT by Race group
# Purpose: Test if mean ALT differs across race groups.
# RIDRETH1: Race/Hispanic origin
# -------------------------------------------------

anova_data = df_all[['ALT (U/L)', 'RIDRETH1']].dropna()
groups = [anova_data['ALT (U/L)'][anova_data['RIDRETH1'] == g] for g in anova_data['RIDRETH1'].unique()]
f_stat, p_val = stats.f_oneway(*groups)
print(f"ANOVA ALT by Race: F-stat = {f_stat:.2f}, p = {p_val:.4f}")

# -------------------------------------------------
# 📌 3️⃣ Multiple Linear Regression
# Purpose:
# Predict ALT using:
# - BMI
# - Sleep Hours
# - Alcohol
# - Physical Activity
# - Age
# - Gender
# -------------------------------------------------

# Prepare model data
model_data = df_all[['ALT (U/L)', 'BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use', 'Physical Activity', 'Age', 'RIAGENDR']].dropna()

# Define predictors and outcome
X = model_data[['BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use', 'Physical Activity', 'Age', 'RIAGENDR']]
X = sm.add_constant(X)
y = model_data['ALT (U/L)']

# Fit OLS regression
model = sm.OLS(y, X).fit()
print(model.summary())

# -------------------------------------------------
# 📌 Optional: Stratified Analysis
# Purpose:
# Repeat regression for Males and Females separately.
# -------------------------------------------------

# Males only
male_data = model_data[model_data['RIAGENDR'] == 1]
X_m = male_data[['BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use', 'Physical Activity', 'Age']]
X_m = sm.add_constant(X_m)
y_m = male_data['ALT (U/L)']

model_male = sm.OLS(y_m, X_m).fit()
print("\n--- Regression: Males only ---")
print(model_male.summary())

# Females only
female_data = model_data[model_data['RIAGENDR'] == 2]
X_f = female_data[['BMI (kg/m2)', 'Sleep (hrs)', 'Alcohol Use', 'Physical Activity', 'Age']]
X_f = sm.add_constant(X_f)
y_f = female_data['ALT (U/L)']

model_female = sm.OLS(y_f, X_f).fit()
print("\n--- Regression: Females only ---")
print(model_female.summary())

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np # Import numpy for isnan check and where

# --- Print columns for debugging ---
print("Columns in df_all:", df_all.columns)

# --- Define thresholds for abnormal ---
ALT_thresh = 40
AST_thresh = 40
GGT_thresh_male = 51
GGT_thresh_female = 33

# --- Add abnormal flag columns ---
# Use the renamed columns 'ALT (U/L)', 'AST (U/L)', 'GGT (U/L)'
df_all["ALT_abnormal"] = df_all["ALT (U/L)"] > ALT_thresh
df_all["AST_abnormal"] = df_all["AST (U/L)"] > AST_thresh

# Calculate GGT_abnormal using numpy.where for better handling of conditions
# Check if RIAGENDR is 1 (Male) and GGT > male threshold
is_male_abnormal_ggt = (df_all["RIAGENDR"] == 1) & (df_all["GGT (U/L)"] > GGT_thresh_male)

# Check if RIAGENDR is 2 (Female) and GGT > female threshold
is_female_abnormal_ggt = (df_all["RIAGENDR"] == 2) & (df_all["GGT (U/L)"] > GGT_thresh_female)

# GGT_abnormal is True if either male or female condition is met
df_all["GGT_abnormal"] = np.where(
    is_male_abnormal_ggt | is_female_abnormal_ggt,
    True,
    False
)

# --- Create categorical variables for stratification ---

# Age groups (example)
bins_age = [0, 20, 40, 60, 80, 120]
labels_age = ["0-19", "20-39", "40-59", "60-79", "80+"]
df_all["AgeGroup"] = pd.cut(df_all["Age"], bins=bins_age, labels=labels_age, right=False)

# Race mapping (example for RIDRETH1)
race_map = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    5: "Other Race - Including Multi-Racial"
}
df_all["Race"] = df_all["RIDRETH1"].map(race_map)

# Alcohol use binary: 1 = Yes, 2 = No (example ALQ101: Ever had 12+ drinks lifetime)
# Check if 'Alcohol Use' column exists and map values. Assuming 1=Yes, 2=No from previous data.
if 'Alcohol Use' in df_all.columns:
    df_all["AlcoholUse"] = df_all["Alcohol Use"].map({1: "Yes", 2: "No"})
else:
    print("Warning: 'Alcohol Use' column not found for AlcoholUse mapping.")
    df_all["AlcoholUse"] = None # Or handle as appropriate


# BMI categories
bins_bmi = [0, 18.5, 25, 30, 100]
labels_bmi = ["Underweight", "Normal", "Overweight", "Obese"]
df_all["BMICategory"] = pd.cut(df_all["BMI (kg/m2)"], bins=bins_bmi, labels=labels_bmi, right=False)

# Sleep duration categories (example from SLD010H)
# Use the renamed 'Sleep (hrs)' column. Define bins based on common recommendations.
bins_sleep = [0, 6, 7, 9, 24] # Adjusted bins for more common categories
labels_sleep = ["Short", "Recommended", "Long", "Very Long"]
df_all["SleepCategory"] = pd.cut(df_all["Sleep (hrs)"], bins=bins_sleep, labels=labels_sleep, right=False)


# Physical activity binary example (PAQ650: Yes=1, No=2)
# Use the renamed 'Physical Activity' column. Assuming 1=Active, 2=Inactive based on previous data.
if 'Physical Activity' in df_all.columns:
    df_all["PhysActivity"] = df_all["Physical Activity"].map({1: "Active", 2: "Inactive"})
else:
    print("Warning: 'Physical Activity' column not found for PhysActivity mapping.")
    df_all["PhysActivity"] = None # Or handle as appropriate


# --- Define a helper function to calculate % abnormal ---
def calc_abnormal_percent(df, group_col, abnormal_col):
    # Drop rows with missing values in the grouping column
    df_cleaned = df.dropna(subset=[group_col])
    grouped = df_cleaned.groupby(group_col)[abnormal_col].mean() * 100
    return grouped.reset_index().rename(columns={abnormal_col: "Percentage"})

# --- Plot function ---
def plot_abnormal_by_group(df, group_col, enzyme_name):
    # Check if the grouping column exists and is not all missing
    if group_col not in df.columns or df[group_col].isnull().all():
        print(f"Skipping plot for {enzyme_name} by {group_col}: Column not found or all missing.")
        return

    df_plot = calc_abnormal_percent(df, group_col, enzyme_name+"_abnormal")
    plt.figure(figsize=(10,6))
    sns.barplot(data=df_plot, x=group_col, y="Percentage")
    plt.title(f"Percentage with Abnormal {enzyme_name} by {group_col}")
    plt.ylabel("Percentage (%)")
    plt.xlabel(group_col)
    plt.xticks(rotation=45)
    plt.ylim(0, 100)
    plt.tight_layout()
    plt.show()

# --- Example plots ---
# Ensure that the columns used for stratification actually exist in the dataframe
strat_vars = ["RIAGENDR", "AgeGroup", "Race", "AlcoholUse", "BMICategory", "SleepCategory", "PhysActivity"]
available_strat_vars = [v for v in df_all.columns if v in strat_vars]

for enzyme in ["ALT", "AST", "GGT"]:
    for strat_var in available_strat_vars:
        print(f"Plotting {enzyme} abnormal by {strat_var} ...")
        plot_abnormal_by_group(df_all, strat_var, enzyme)
```
